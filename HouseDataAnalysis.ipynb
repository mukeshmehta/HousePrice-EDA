```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
df_train = pd.read_csv("train.csv") #Training Data
```

```python
df_train.columns
```

```python
df_train['SalePrice'].describe()
```

```python
sns.distplot(df_train['SalePrice'])
```

```python
df_train['SalePrice'].skew()  #Skewness
```

```python
df_train['SalePrice'].kurt()  #Kurtosis
```

## SalePrice vs. GrLivArea

```python
data1 = pd.concat([df_train['SalePrice'], df_train['GrLivArea']], axis=1)
data1.plot.scatter(x='GrLivArea', y='SalePrice', xlim=(0,6000), ylim=(0,800000));
```

It looks like SalePrice and GrLivArea are having <b>Linear Relationship </b>

### TotalBsmtSF Vs. SalePrice

```python
data2 = pd.concat([df_train['SalePrice'], df_train['TotalBsmtSF']], axis=1)
data2.plot.scatter(x='TotalBsmtSF', y='SalePrice', xlim=(0,6000), ylim=(0,800000));
```

SalePrice and TotalBsmt are having <b>(Strong) Linear Relationship </b>

### Relationship with Categorical Features

```python
data3 = pd.concat([df_train['SalePrice'],df_train['OverallQual']], axis = 1)
f, ax = plt.subplots(figsize=(8,6))
fig = sns.boxplot(x = 'OverallQual', y='SalePrice', data = data3)
fig.axis(ymin = 0, ymax= 800000)
```

```python
data4 = pd.concat([df_train['SalePrice'],df_train['YearBuilt']], axis = 1)
fig = sns.boxplot(x = 'YearBuilt', y = 'SalePrice', data = data4)
```

### Correlation Matrix

```python
corrmat = df_train.corr()
f, ax = plt.subplots(figsize = (14,10))
sns.heatmap(corrmat, vmax = 0.8, square = True)
```
